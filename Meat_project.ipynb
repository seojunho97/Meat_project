```python
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
from selenium import webdriver
```

```python
url_meat_breed = 'http://www.kmta.or.kr/kr/data/stats_breed_beef.php'
driver = webdriver.Chrome('chromedriver')
driver.get(url_meat_breed)
```

```python
data = driver.find_elements_by_css_selector('body > div.container > div.contents.p_stats > table > tbody > tr')
```

```python
breed_datas = []
for i in data[2:-1]:
    text = i.text
    breed_data = text.split()
    breed_datas.append(breed_data)
print(breed_datas)
```
